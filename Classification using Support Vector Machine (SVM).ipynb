```python
import numpy as np
import pandas as pd
```

```python
dataFile='/pluralsight/ad-dataset/ad.data'
data=pd.read_csv(dataFile, sep=",",header=None, low_memory=False)
```

```python
data.head(20)
```

```python
def toNum(cell):
    try:
        return np.float(cell)
    except:
        return np.nan

def seriestoNum(series):
    return series.apply(toNum)
```

```python
train_data=data.iloc[0:,0:-1].apply(seriestoNum)
```

```python
train_data.head(20)
```

```python
train_data=train_data.dropna()
train_data.head(20)
```

```python
def toLabel(str):
    if str=="ad.":
        return 1
    else:
        return 0
```

```python
train_labels=data.iloc[train_data.index,-1].apply(toLabel)
train_labels
```

```python
from sklearn.svm import LinearSVC

clf = LinearSVC()
clf.fit(train_data[100:2300], train_labels[100:2300])
```

```python
#Test Phase
clf.predict(train_data.iloc[12].values.reshape(1,-1))
```

```python
#Test Phase
clf.predict(train_data.iloc[-1].values.reshape(1,-1))
```
